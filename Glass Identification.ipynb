```python
x=df1.drop('Type of glass',axis=1)
y=df1["Type of glass"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['VIF Values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['FEATURES']=x.columns
vif
```

VIF value of Magnesium, Silicon, Calcium is high & these columns has high correlation. We have very less number of rows so lets proceed with same vif values.

```python
y.value_counts()      # The data was immbalace so lets do Oversampling by using smote method.
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()          # Now the data become balanced.
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
# Finding the best random state
maxacc=0
maxrs=0
for i in range(1,420):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=.20)
    rfc=RandomForestClassifier()
    rfc.fit(x_train,y_train)
    predict_rfc=rfc.predict(x_test)
    acc=accuracy_score(y_test,predict_rfc)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print("Best accuracy score is ",maxacc,"at random state ",maxrs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=maxrs,test_size=.20)
```

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predict_RFC=RFC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_RFC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_RFC))
print("Classification Report : ",classification_report(y_test,predict_RFC))
```

```python
ETC=ExtraTreesClassifier()
ETC.fit(x_train,y_train)
predict_ETC=ETC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_ETC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_ETC))
print("Classification Report : ",classification_report(y_test,predict_ETC))
```

```python
GBC=GradientBoostingClassifier()
GBC.fit(x_train,y_train)
predict_GBC=GBC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_GBC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_GBC))
print("Classification Report : ",classification_report(y_test,predict_GBC))
```

```python
ABC=AdaBoostClassifier()
ABC.fit(x_train,y_train)
predict_ABC=ABC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_ABC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_ABC))
print("Classification Report : ",classification_report(y_test,predict_ABC))
```

```python
BC=BaggingClassifier()
BC.fit(x_train,y_train)
predict_BC=BC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_BC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_BC))
print("Classification Report : ",classification_report(y_test,predict_BC))
```

```python
KNC=KNeighborsClassifier()
KNC.fit(x_train,y_train)
predict_KNC=KNC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_KNC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_KNC))
print("Classification Report : ",classification_report(y_test,predict_KNC))
```

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
predict_DTC=DTC.predict(x_test)
print("Accuracy score : ",accuracy_score(y_test,predict_DTC))
print("Confusion Matrix : ",confusion_matrix(y_test,predict_DTC))
print("Classification Report : ",classification_report(y_test,predict_DTC))
```

```python
final_model=ExtraTreesClassifier(criterion='entropy',n_estimators=45,random_state=389)
final_model.fit(x_train,y_train)
pred_fin_model=final_model.predict(x_test)               # Random Forest classifier working wel
acc=accuracy_score(y_test,pred_fin_model)
print(acc*100)
```

```python
y_pred_proba=final_model.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_proba,pos_label=1)            # Auc Roc curve 
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Random Forest Classifier') 
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
auc_score=roc_auc_score(y_test,final_model.predict(x_test),multi_class='ovo')
print("Auc score : ",auc_score)
```
